```python
import GEOparse
import pandas as pd

# Function to fetch gene expression data from GEO
def fetch_geo_data(gse_id):
    gse = GEOparse.get_GEO(geo=gse_id, destdir="./")
    # Extract expression data
    expression_data = gse.pivot_samples('VALUE')
    # Extract gene symbol annotation
    gene_symbols = gse.gpls[gse.gpls.keys()[0]].table[['ID', 'Gene Symbol']]
    gene_symbols.columns = ['ID_REF', 'Gene Symbol']
    return expression_data, gene_symbols



# Fetch gene expression data (replace 'GSE12345' with your actual GEO Series ID)
gse_id = 'GSE12345'  # replace with your actual GEO Series ID
expression_data, gene_symbols = fetch_geo_data(gse_id)

# Reshape the expression data to long format for merging
expression_data = expression_data.reset_index().melt(id_vars=['ID_REF'], var_name='GEO Accession:', value_name='Expression Value')

# Merge expression data with gene symbols
expression_data = pd.merge(expression_data, gene_symbols, on='ID_REF')

# Merge the dataframes on 'GEO Accession:' and 'Gene Symbol'
merged_expression_data = pd.merge(merged_data, expression_data, on=['GEO Accession:', 'Gene Symbol'], how='left')

# Extract relevant columns
extracted_data = merged_expression_data[['GEO Accession:', 'Sample Name:', 'Gene Symbol', 'Expression Value']]

# Display all rows of the extracted data
pd.set_option('display.max_rows', None)
print(extracted_data)
```

### B . Differentially Expressed Genes

```python
import numpy as np
from scipy import stats

# Convert gene expression values to numeric type
merged_data['Gene Symbol'] = pd.to_numeric(merged_data['Gene Symbol'], errors='coerce')

# Step 1: Separate gene expression values for AD and non-AD samples
ad_samples = merged_data.loc[merged_data['Disease State:'] == 'Alzheimer\'s Disease', 'Gene Symbol']
non_ad_samples = merged_data.loc[merged_data['Disease State:'] == 'normal', 'Gene Symbol']

# Step 2: Calculate Mean Expression
mean_ad = ad_samples.mean()
mean_non_ad = non_ad_samples.mean()

# Step 3: Perform Welch's t-test
t_stat, p_value = stats.ttest_ind(ad_samples, non_ad_samples, equal_var=False)

# Step 4: Filter Genes
significant_genes = p_value < 0.05

# Step 5: Calculate Fold Change
fold_change = mean_ad - mean_non_ad

# Print or use the results as needed
print("Significant Genes:", significant_genes)
print("Fold Change:", fold_change)
```

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


# Drop any non-numeric columns or set them aside
# Assuming the gene expression data is in all columns of merged_data
gene_data = merged_data.select_dtypes(include=[np.number])

# Standardizing the data before PCA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_gene_data = scaler.fit_transform(gene_data)

# Apply PCA
pca = PCA()
pca.fit(scaled_gene_data)

# Transform the data to get the principal components
pca_data = pca.transform(scaled_gene_data)

# Get the component loadings (scores for each gene)
loadings = pd.DataFrame(pca.components_.T, columns=[f'PC{i+1}' for i in range(pca.n_components_)], index=gene_data.columns)

# Select the top scored feature (gene) in each component
top_genes = set()
for pc in loadings.columns:
    top_gene = loadings[pc].abs().idxmax()  # Gene with the highest absolute score in the component
    top_genes.add(top_gene)
    if len(top_genes) >= 50:
        break

# If fewer than 50 unique genes are selected, fill up to 50
if len(top_genes) < 50:
    additional_genes = loadings.abs().mean(axis=1).nlargest(50-len(top_genes)).index
    top_genes.update(additional_genes)

# Extract the selected top genes
selected_genes = list(top_genes)[:50]
print(f'Selected Genes: {selected_genes}')

# Visualizing the explained variance by each principal component
plt.figure(figsize=(10, 7))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by PCA Components')
plt.grid(True)
plt.show()

# If you want to create a new dataframe with only the selected genes
selected_gene_data = merged_data[selected_genes]

# Save the selected genes data if needed
selected_gene_data.to_csv('selected_gene_data.csv', index=False)
```
